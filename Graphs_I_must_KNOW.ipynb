```python
#https://www.youtube.com/watch?v=NrmmR3-VxA8&list=PLDesaqWTN6ESsmwELdrzhcGiRhk5DjwLP&index=13
import sympy as smp
from sympy import *
from sympy.simplify.simplify import nthroot
import numpy as np

import matplotlib as plt
from matplotlib import *
```

**Constant or Horizontal Line$ f(x)=2$**

```python
x = symbols ('x')
x=2
f = (x)
plot(f)
```

**Hyperbola      $f(x) = x^2 $**

```python
x = symbols ('x')
f = (x**2)
rcParams["figure.figsize"] = (10, 10)
plot(f)
```

**Hyperbola transformed by changing       $f(x) = x^2 $ to $f(x) = x^2 + 3$ shifts up the Y axis 3**

```python
x = symbols ('x')
f = (x**2)+ 3
rcParams["figure.figsize"] = (10, 10)
plot(f)
```

**Absolute $ f(x) =|x| $**

```python
x = symbols ('x')
f = abs(x)
plot(f)
```

**Absolute transformed by changing  $f(x) =|x| $ to $ f(x) =|x+2|$ shifts down the X axis 2**

```python
x = symbols ('x')
f = abs(x+2)
plot(f)
```

**Square root$ f(x) = \sqrt x $**

```python
x = symbols ('x')
f = sqrt(x)
plot(f)
```

**Square root  $ f(x) = \sqrt x $  to $ f(x) = (\sqrt x-2)+3 $ (0,0) shifts up 3 and 2 to the right. to (2,3)**

```python
x = symbols ('x')
f = sqrt(x-2)+3
plot(f)
```

**Square root$ f(x) = \sqrt x $**

**Identity: $ f(x) =  x $**

```python
x = symbols ('x')
f = x
plot(f)
```

**Cubic $ f(x) =  x^3 $**

```python
x = symbols ('x')
f = x**3
plot(f)
```

**Cube Root$ f(x) = \sqrt[3] x $**

```python
x = symbols ('x')
f = real_root(x,3)
#cbrt and nthroot both have issues when it comes to negatives here
rcParams["figure.figsize"] = (20, 10)
plot(f)
```

**Recipricol $ f(x) = {\frac {1}{x}} $**

```python
x = symbols ('x')
f = 1/x
plot(f)
```

**Recipricol Squared $ f(x) = {\frac {1}{x^2}} $**

```python
x = symbols ('x')
f = 1/x**2
rcParams["figure.figsize"] = (10, 5)
plot(f)
```
